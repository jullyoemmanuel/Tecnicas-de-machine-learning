```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Redimencionando o tamanho das figuras
plt.rcParams['figure.figsize'] = (15,8)
```

```python
empresas = ['Apple','Google','Amazon',
    'Microsoft','Tencent','McDonalds',
    'Visa','Facebook','Alibaba','Louis Vuitton']
```

```python
bilhoes = [947.062,819.573,705.646,
        611.460,214.023,196.526,
        191.032,186.421,169.966,
        124.273]
```

```python
plt.plot(empresas,bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilh천es')
```

```python
# conhecendo o fontsize
plt.plot(empresas,bilhoes)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh천es', fontsize = 20)
```

```python
# mudando a cor e legendando
plt.plot(empresas,bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh천es', fontsize = 20)
plt.legend(['Bilh천es de d처lores'], loc = 'right', fontsize = 15)
```

```python
# aumentar a espessura da linha lw
plt.plot(empresas,bilhoes, color = 'green', lw = 6)
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh천es', fontsize = 20)
plt.legend(['Bilh천es de d처lores'], loc = 'right', fontsize = 15)
```

```python
# adicionar marcador

plt.plot(empresas,bilhoes, color = 'green', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh천es', fontsize = 20)
plt.legend(['Bilh천es de d처lores'], loc = 'right', fontsize = 15)
```

```python
# adicionando linhas de grade

plt.plot(empresas,bilhoes, color = 'green', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh천es', fontsize = 20)
plt.legend(['Bilh천es de d처lores'], loc = 'right', fontsize = 15)
plt.grid(True)
```

```python
#mudando o tipo do gr찼fico e colocando a legenda acima
plt.bar(empresas,bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh천es', fontsize = 20)
plt.legend(['Bilh천es de d처lores'], loc = 'center right', fontsize = 15)
```

```python
# gr찼fico mais simples 
plt.bar(empresas, bilhoes)
```

```python
cores = ['Black', 'dimgray','dimgrey','gray','grey',
         'darkgray','darkgrey','silver','lightgray','lightgrey']
```

```python
plt.bar(empresas,bilhoes,color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh천es', fontsize = 20)
plt.grid(True)
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
plt.bar(empresas,bilhoes,color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 20)
plt.ylabel('Bilh천es', fontsize = 20)
```

```python
# barh 
plt.style.use('fivethirtyeight')
plt.barh(empresas,bilhoes,color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Bilh천es de d처lares', fontsize = 15)
plt.ylabel('Empresas', fontsize = 15)
```

```python
#scatter
plt.style.use('fivethirtyeight')
plt.scatter(empresas,bilhoes,color = cores, edgecolor = 'black')
plt.title('Empresas mais ricas do mundo', fontsize = 25)
plt.xlabel('Empresas', fontsize = 15)
plt.ylabel('Bilh천es de d처lares', fontsize = 15)
```
